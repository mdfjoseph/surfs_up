```python
# Dependencies
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# Import the sqlalchemy extract function.
from sqlalchemy import extract

# 1. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_results = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6)
```

```python
# 2. Convert the June temperatures to a list.
june_results = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
print(june_results)
```

```python
# 3. Create a DataFrame from the list of temperatures for the month of June. 
june_results_df = pd.DataFrame(june_results, columns=['date', 'June Temps'])
june_results_df
```

```python
# 4. Calculate and print out the summary statistics for the June temperature DataFrame.
print(june_results_df.describe().to_string(index=True))
june_results_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_results = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 12)
```

```python
# 7. Convert the December temperatures to a list.
december_results = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
print(december_results)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_results_df = pd.DataFrame(december_results, columns=['date', 'December Temps'])
december_results_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
print(december_results_df.describe().to_string(index=True))
december_results_df.describe()
```

```python
# 10. Return to your challenge instructions to compare your summary statistics
june_statistics = june_results_df.describe()
dec_statistics = december_results_df.describe()

stat_summary_df = june_statistics.merge(dec_statistics, left_index=True, right_index=True)
stat_summary_df
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
stat_summary_df.plot()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
december_results_df.plot.hist(bins=20)
plt.tight_layout()
```
